```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# drop rows with missing Age values
df.dropna(subset=['Age'], inplace=True)

# Define features and target for regression
features = ['Age', 'Pclass']
target_reg = 'Fare'

X_reg = df[features]
y_reg = df[target_reg]

# training and testing sets data split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

print("Regression training data shape:", X_train_reg.shape)
print("Regression testing data shape:", X_test_reg.shape)
print("\nFirst five rows of regression features:")
print(X_reg.head())
```

```python
# Task 1

# model instance
lr_model = LinearRegression()

# training the model
lr_model.fit(X_train_reg, y_train_reg)

# predictions
y_pred_reg = lr_model.predict(X_test_reg)

# evaluation
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)

print(f"Mean Squared Error for Fare Prediction: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")

print("\nLinear Regression Coefficients:")
for feature, coefficient in zip(features, lr_model.coef_):
    print(f"{feature}: {coefficient:.2f}")

print(f"\nIntercept: {lr_model.intercept_:.2f}")

print("\nSample predictions:")
comparison_df = pd.DataFrame({
    'Actual Fare': y_test_reg.values,
    'Predicted Fare': y_pred_reg
})

print(comparison_df.head(10))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Prepare data for classification
# Logistic Regression predicts a category, so here we are predicting Survived.

# encode the Sex column. models need numbers
df['Sex'] = df['Sex'].map({
    'male': 0,
    'female': 1
})

# Define features and target for classification
features_cls = ['Age', 'Pclass', 'Sex']
target_cls = 'Survived'

X_cls = df[features_cls]
y_cls = df[target_cls]

# Split the data into training and testing sets
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42
)

print("Classification training data shape:", X_train_cls.shape)
print("Classification testing data shape:", X_test_cls.shape)
print("\nFirst five rows of classification features:")
print(X_cls.head())
```

```python
# Task 2

# model instance
log_model = LogisticRegression(random_state=42)

# train the model
log_model.fit(X_train_cls, y_train_cls)

# predictions
y_pred_cls = log_model.predict(X_test_cls)

# evaluation
accuracy = accuracy_score(y_test_cls, y_pred_cls)

print(f"Accuracy for Survival Prediction: {accuracy:.2%}")

print("\nClassification Report:")
print(classification_report(y_test_cls, y_pred_cls))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_cls, y_pred_cls))

print("\nLogistic Regression Coefficients:")
for feature, coefficient in zip(features_cls, log_model.coef_[0]):
    print(f"{feature}: {coefficient:.2f}")

print(f"\nIntercept: {log_model.intercept_[0]:.2f}")
```

**Knowledge Check**

1. In your own words, what is the key difference between a regression problem and a classification problem?

    - The main difference is that regression predicts a number, while classification predicts a category. In this lab, predicting Fare is regression because fare is a continuous dollar amount, but predicting Survived is classification because the answer is either 0 or 1.

2. The LinearRegression model has an attribute called .coef_. After you train the model, print lr_model.coef_. What do these numbers represent?

    - The .coef_ values show how much each feature affects the model's prediction. For the Linear Regression model, the coefficients explain how Age and Pclass are connected to the predicted fare. For example, if a coefficient is positive, the predicted fare goes up as that feature increases. If it is negative, the predicted fare goes down as that feature increases.

3. Why did we use mean_squared_error to evaluate the regression model but accuracy_score for the classification model? Why wouldn't accuracy be a good metric for the fare prediction task?

    - We used mean_squared_error for the regression model because fare prediction is a number-based problem, so we need to measure how far off the predicted fare was from the actual fare. We used accuracy_score for the classification model because survival prediction is a category problem, so we can count how many predictions were correct. Accuracy would not be good for fare prediction because fare is not simply right or wrong like a class label; a prediction can be close, far off, or somewhere in between.
